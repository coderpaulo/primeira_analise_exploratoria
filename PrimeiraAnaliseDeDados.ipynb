#Importando e limpando os dados:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('vgsales.csv')
data.columns= data.columns .str.capitalize()
data.head()
```

```python
data.columns = ['Rank', 'Nome', 'Plataforma', 'Ano', 'Genêro', 'Editora', 'Vendas_America_do_Norte_Vendas',
       'Vendas_Europa', 'Vendas_Japão', 'Outras_Vendas', 'Vendas_Globais']
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
#somando todos os valores nulos
```

```python
data.dropna(inplace=True)         #excluindo valores nulos.
data.shape[0]
```

```python
data['Ano']=data['Ano'].astype(int)     #convertendo de float para int
```

```python
data.duplicated().sum()
```

```python
data.nunique()
```

```python
data.describe()
```

# Analise de vendas por gênero:

 Qual genêro tem maior numero de jogo?

```python
data['Genêro'].value_counts()
```

```python
bars = data.Genêro.value_counts()
bars.plot(kind="bar",title='Gêneros',figsize=(8,5))
plt.show()
```

```python
vendas_por_genero = data.groupby('Genêro')['Vendas_Globais'].sum().sort_values(ascending=False)
vendas_por_genero.plot(kind="bar",title='Vendas por genêro',figsize=(8,5))
```

# Analise de vendas por plataforma:

```python
data.Plataforma.value_counts()
```

```python
bars = data.Plataforma.value_counts()
bars.plot(kind="bar",title='Plataformas',figsize=(8,5))
plt.show()
```

Análise de vendas globais por plataforma:

```python
vendas_por_plataforma = data.groupby('Plataforma')['Vendas_Globais'].sum().sort_values(ascending=False)
vendas_por_plataforma.plot(kind="bar",title='Plataformas',figsize=(8,5))
plt.show()
```

Analise de vendas de gênero por plataforma:

```python
data.groupby(['Plataforma', 'Genêro'])['Vendas_Globais'].sum().unstack().plot(kind='bar', stacked=True, figsize=(13, 6))
```

# Editoras de maior sucesso:

```python
data.Editora.unique()
```

```python
data.Editora.value_counts().index
```

```python
contagem_editoras = data.Editora.value_counts().to_frame().reset_index()
contagem_editoras.columns = ["Editoras", "Total"]
contagem_editoras
```

```python
grupo_editoras = data.groupby('Editora')
vendas_por_editora = grupo_editoras['Vendas_Globais'].sum()
vendas_por_editora = vendas_por_editora.sort_values(ascending=False)
top_10_editoras = vendas_por_editora.head(10)
```

```python
plt.figure(figsize=(10, 6))
top_10_editoras.plot(kind='bar')
plt.title('Top 10 Editoras de Jogos por Vendas Globais')
plt.ylabel('Vendas Globais (em milhões)')
plt.show()
```

# Analise geral sob vendas de jogos:

Encontrando os jogos lançados em ordem decrescente de anos:

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="Ano", data=data, order=data.groupby(by=['Ano'])['Nome'].count().sort_values(ascending=False).index)
# Cria um gráfico de contagem (countplot) com o eixo x sendo os anos e os dados vindo do DataFrame
plt.xticks(rotation='vertical')
plt.title('Jogos Lançados por Ano')
plt.ylabel('Contagem')
plt.xlabel('Ano')
```

# Analise de vendas por região:

```python
região = ['Vendas_America_do_Norte_Vendas', 'Vendas_Europa', 'Vendas_Japão', 'Outras_Vendas']
vendas_por_região = data[região].sum()
```

```python
plt.figure(figsize=(10, 6))
vendas_por_região.plot(kind='bar', color=['blue', 'green', 'red', 'orange'])
plt.xlabel('Região')
plt.ylabel('Total de vendas (em milhões)')
plt.title('Total de vendas por região')
plt.xticks(rotation=0)
plt.show()
```

Qual ano tem maior numero de vendas no mundo todo?

```python
vendas_por_ano = data.groupby('Ano')['Vendas_Globais'].sum()
ano_de_maior_venda = vendas_por_ano.idxmax()
print(ano_de_maior_venda)
```

```python
ano_vendas_de_video_games = data.groupby('Ano')[['Vendas_Globais','Vendas_America_do_Norte_Vendas','Vendas_Europa','Vendas_Japão','Outras_Vendas']].sum()
ano_vendas_de_video_games.plot(figsize = (12,8))
plt.axvline(x=ano_de_maior_venda ,c = 'black',linestyle = '-.',alpha = 0.6)
plt.ylabel('Vendas (milhões de dolares)',fontsize = 12)
plt.show()
```
